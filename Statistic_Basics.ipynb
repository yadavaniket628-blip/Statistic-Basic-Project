1. **What is the difference between descriptive statistics and inferential
statistics?**  **Explain with examples.**



**Descriptive Statistics** -

1.   It gives information about raw data which describes the data in some manner.
2.   It helps in organizing, analyzing, and to present data in a meaningful manner.
3.   It is used to describe a situation.
4.   It explains already known data and is limited to a sample or population having a small size.
5.  ** Examples include**: mean, median, mode, range, variance, histograms, pie charts.
6.   Limited to presenting and analyzing known data.


**Inferential Statistics**

1.   It makes inferences about the population using data drawn from the population.
2.   It allows us to compare data, and make hypotheses and predictions.
3.   It is used to explain the chance of occurrence of an event.
4.   It attempts to reach the conclusion about the population.
5.   **Examples include**: confidence intervals, hypothesis testing, regression models, p-values.
6.   Allows predictions and conclusions that go beyond the data at hand.


**Descriptive Statistic** Example :-

Example: Amazon Seller Revenue Analysis
Suppose you collect monthly revenue data from 50 Amazon sellers. Here's how descriptive statistics help:
1. Measures of Central Tendency
- Mean (Average): ₹80,000 — the average monthly revenue across all sellers.
- Median: ₹75,000 — the middle value when revenues are sorted.
- Mode: ₹70,000 — the most frequently occurring revenue.
2. Measures of Dispersion
- Range: ₹40,000 to ₹1,20,000 — shows the spread between lowest and highest revenue.
- Standard Deviation: ₹15,000 — tells you how much the revenues vary from the average.
3. Visualizations
- Histogram: Shows how many sellers fall into each revenue bracket.
- Pie Chart: Displays the percentage of sellers in different product categories.

**Inferential statistics** Example:-

- Average revenue = ₹80,000

2. **What is sampling in statistics? Explain the differences between random
and stratified sampling.**

Sampling is the statistical process of selecting a subset of a population to analyze, allowing researchers to estimate the characteristics of the entire group without collecting data from every single member. The goal is to create a representative sample that can be used to make valid inferences about the larger population, which is often more cost-effective and faster than a full census.


**Random sampling selects individuals purely by chance, while stratified sampling divides the population into subgroups and samples proportionally from each.**

Here’s a detailed breakdown of how they differ:

---

###  **Simple Random Sampling**
- **Definition**: Every individual in the population has an *equal chance* of being selected.
- **Method**: Selection is done entirely by chance, often using random number generators or lottery methods.
- **Use Case**: Best when the population is *homogeneous* or when subgroup representation isn’t critical.
- **Advantages**:
  - Easy to implement.
  - Minimizes selection bias.
- **Disadvantages**:
  - May not represent key subgroups if the population is diverse.
  - Can lead to sampling error if important segments are underrepresented.

---

###  **Stratified Sampling**
- **Definition**: The population is divided into *strata* (distinct subgroups) based on shared characteristics (e.g., age, income, location), and samples are drawn from each stratum.
- **Method**: Each stratum is sampled randomly, either proportionally or equally.
- **Use Case**: Ideal when the population has *distinct subgroups* and you want to ensure each is adequately represented.
- **Advantages**:
  - Ensures representation of all key subgroups.
  - Reduces sampling error and increases precision.
- **Disadvantages**:
  - More complex to design and execute.
  - Requires detailed population data to define strata.




3. **Define mean, median, and mode. Explain why these measures of central
tendency are important.**


**Mean, median, and mode are statistical tools used to identify the central or typical value in a dataset. They help summarize and interpret data effectively, guiding decisions in fields like business, economics, and research.**

---
 **Definitions of Mean, Median, and Mode**

- **Mean (Arithmetic Average)**  
  - Calculated by adding all values in a dataset and dividing by the number of values.  
  - Example: For scores 10, 20, 30 → Mean = (10 + 20 + 30)/3 = 20  
  - **Sensitive to outliers**, which can skew the result.

- **Median (Middle Value)**  
  - The value that lies in the center when data is arranged in ascending or descending order.  
  - Example: For scores 10, 20, 30 → Median = 20  
  - If the dataset has an even number of values, the median is the average of the two middle values.  
  - **Resistant to outliers**, making it useful for skewed distributions.

- **Mode (Most Frequent Value)**  
  - The value that appears most often in a dataset.  
  - Example: For scores 10, 20, 20, 30 → Mode = 20  
  - A dataset can have **no mode**, **one mode**, or **multiple modes** (bimodal/multimodal).

---
 **Why These Measures Matter**

- **Simplify complex data**: They reduce large datasets to a single representative value.
- **Support decision-making**: Businesses use the mean to calculate average sales, median to understand income distribution, and mode to identify popular products.
- **Reveal data distribution**: Comparing mean and median can indicate skewness (e.g., income data often has a higher mean than median).
- **Tailor analysis to context**:
  - Use **mean** for normally distributed data.
  - Use **median** when data has outliers or is skewed.
  - Use **mode** for categorical data or identifying trends.

---

These measures are foundational in statistics and data analysis. Whether you're evaluating seller performance, customer behavior, or pricing strategies, choosing the right measure of central tendency ensures your insights are accurate and actionable.





4. **Explain skewness and kurtosis. What does a positive skew imply about the
data?**

**Skewness**: Measuring Asymmetry
- Definition: Skewness quantifies how much a distribution deviates from symmetry.
- Types:
- Symmetric (Skewness = 0): Mean = Median = Mode. The distribution is balanced on both sides.
- Positive Skew (Right-Skewed): The right tail is longer. Most data points are concentrated on the left. Mean > Median > Mode.
- Negative Skew (Left-Skewed): The left tail is longer. Most data points are on the right. Mean < Median < Mode.

 **Kurtosis**: Measuring Tailedness
- Definition: Kurtosis describes the "tailedness" of a distribution—how heavy or light the tails are compared to a normal distribution.
- Types:
- Mesokurtic (Kurtosis ≈ 3): Normal distribution; moderate tails.
- Leptokurtic (Kurtosis > 3): Heavy tails; more outliers than normal.
- Platykurtic (Kurtosis < 3): Light tails; fewer outliers.
High kurtosis indicates a higher likelihood of extreme values (outliers), which can impact statistical models and risk assessments.


A positively skewed distribution indicates that the majority of data points are concentrated on the lower end, with a small number of high-value outliers, creating a long tail on the right side of the graph. This means the mean is greater than the median, which is greater than the mode, because the extreme high values pull the average (mean) to the right.

5. **Implement a Python program to compute the mean, median, and mode of
a given list of numbers.**
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

```python
import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean = statistics.mean(numbers)
median = statistics.median(numbers)
mode = statistics.mode(numbers)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
```

6. **Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:**
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Compute covariance
covariance = np.cov(list_x, list_y)[0][1]

# Compute correlation coefficient
correlation_coefficient = np.corrcoef(list_x, list_y)[0][1]

print(f"Covariance: {covariance}")
print(f"Correlation Coefficient: {correlation_coefficient}")
```

7. **Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:**
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=data)
plt.title('Boxplot of Data')
plt.xlabel('Values')
plt.show()

# Identify outliers using the Interquartile Range (IQR) method
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"\nQ1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR: {IQR}")
print(f"Lower Bound (Q1 - 1.5*IQR): {lower_bound}")
print(f"Upper Bound (Q3 + 1.5*IQR): {upper_bound}")
print(f"Identified Outliers: {outliers}")
```

8. **You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and daily sales.**

● Explain how you would use covariance and correlation to explore this
relationship.

● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

```python
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Compute the correlation coefficient
correlation_coefficient = np.corrcoef(advertising_spend, daily_sales)[0][1]

print(f"Correlation Coefficient between Advertising Spend and Daily Sales: {correlation_coefficient}")
```

**Exploring the Relationship Between Advertising Spend and Daily Sales**

As a data analyst, you can use covariance and correlation to investigate the relationship between advertising spend and daily sales in an e-commerce company.

**Covariance:**

*   **What it measures:** Covariance measures the directional relationship between two variables. A positive covariance indicates that as one variable increases, the other tends to increase as well. A negative covariance indicates that as one variable increases, the other tends to decrease. A covariance close to zero suggests a weak or no linear relationship.
*   **How to use it:** You would calculate the covariance between the `advertising_spend` and `daily_sales` datasets. A positive covariance would suggest that higher advertising spend is associated with higher daily sales.
*   **Limitation:** The magnitude of covariance is not standardized, making it difficult to compare the strength of the relationship across different pairs of variables that are on different scales.

**Correlation Coefficient:**

*   **What it measures:** The correlation coefficient (specifically the Pearson correlation coefficient) measures both the direction and the strength of the linear relationship between two variables. It is a standardized value between -1 and +1.
    *   A value of +1 indicates a perfect positive linear relationship.
    *   A value of -1 indicates a perfect negative linear relationship.
    *   A value of 0 indicates no linear relationship.
    *   Values between 0 and +1 or 0 and -1 indicate the strength of the linear relationship (closer to 1 or -1 means a stronger relationship).
*   **How to use it:** You would calculate the correlation coefficient between `**advertising spend** and **daily sales**. A positive correlation coefficient would indicate that as advertising spend increases, daily sales also tend to increase. The closer the value is to +1, the stronger this positive linear relationship is. This gives you a more interpretable measure of the strength of the association compared to covariance.


9. **Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.**

● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.

● Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

```python
import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(survey_scores, bins=range(4, 12), align='left', rwidth=0.8, color='skyblue', edgecolor='black')
plt.xticks(range(4, 11))
plt.title('Distribution of Customer Satisfaction Scores')
plt.xlabel('Satisfaction Score (1-10)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## Analyzing Customer Satisfaction Survey Data

To understand the distribution of customer satisfaction survey data (on a scale of 1-10) before launching a new product, you can use a combination of summary statistics and visualizations.

**Summary Statistics:**

*   **Mean:** The average satisfaction score. This gives you a single number representing the typical satisfaction level.
*   **Median:** The middle satisfaction score when the data is ordered. This is useful if the data is skewed, as it's less affected by extreme scores than the mean.
*   **Mode:** The most frequent satisfaction score. This tells you which score was given most often by customers.
*   **Standard Deviation:** A measure of the spread or variability of the scores. A higher standard deviation indicates that the scores are more spread out from the mean, while a lower standard deviation indicates they are clustered closer to the mean.
*   **Minimum and Maximum:** The lowest and highest scores in the dataset. This gives you the range of satisfaction levels.

**Visualizations:**

*   **Histogram:** A histogram is an excellent way to visualize the distribution of numerical data. It shows the frequency of each score or range of scores, allowing you to see the shape of the distribution, identify the most common scores, and spot any potential outliers or unusual patterns.
*   **Boxplot:** A boxplot can provide a quick summary of the distribution, including the median, quartiles, and potential outliers.

By using these summary statistics and visualizations, you can gain a good understanding of how satisfied your customers are, identify any areas of concern, and make informed decisions before launching your new product.
